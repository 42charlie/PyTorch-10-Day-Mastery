# Pipeline: Batch Normalization (Vectorized)
**Goal:** Implement a zero-loop normalization pipeline for ImageNet data using Broadcasting.

```python
raw_image = torch.rand(3, 100, 100)
#adding the batch dimension
batched_image = raw_image.reshape(1, 3, 100, 100)
#reshape the means so they can stretch across the entire image via broadcasting.
channels_means = torch.tensor([0.485, 0.456, 0.406]).reshape(1, 3, 1, 1)
normalized_image = batched_image - channels_means
print(f"Normalized Shape: {normalized_image.shape}")
```
