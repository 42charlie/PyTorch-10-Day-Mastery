# Linear Algebra: Slicing & Dot Products
**Goal:** Isolate feature quadrants and apply linear transformations without broadcasting errors.

```python
input_matrix = torch.rand(5, 4)
weights = torch.rand(4, 2)
# select first 2 rows and the last 2 columns
top_right_quadrant = input_matrix[:2, -2:]
result_tensor = input_matrix @ weights
# top_right_quadrant (2, 2) * weights (4, 2) fails the broadcasting rule 
# because 2 != 4 and neither dimension is 1.
print(f"top_right_quadrant: {top_right_quadrant.shape}")
print(f"result_tensor: {result_tensor.shape}")
```
