# Data Stability: Reductions & Types
**Goal:** Calculate dimension-specific statistics while preventing integer/float type crashes.

```python
integer_data = torch.randint(0, 10, [10, 5])
#convert to float because 'mean' cannot operate on ints.
#mean across rows (dim=1) & keepdim=True maintains shape for broadcasting (10, 1).
row_means = torch.mean(integer_data.float(), dim=1, keepdim=True)
print(f" row_means: {row_means}")
print(f" row_means.shape: {row_means.shape}")
```
