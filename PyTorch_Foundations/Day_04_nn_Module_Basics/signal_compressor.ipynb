# The Challenge: The "Signal Compressor"
**Goal:** Build a model that takes a high-dimensional input and progressively "compresses" it down to a single prediction.

```python
class SignalCompressor(torch.nn.Module):
	def __init__(self):
		super().__init__()
		# Defining the layers and the activation object
		self.layer1 = torch.nn.Linear(20, 100)
		self.layer2 = torch.nn.Linear(100, 30)
		self.layer3 = torch.nn.Linear(30, 1)
		self.relu = torch.nn.ReLU()
	
	def forward(self, input_data):
		# Apply layers and activations
		y = self.layer1(input_data)
		y = self.relu(y)
		y = self.layer2(y)
		y = self.relu(y)
		y = self.layer3(y)
		return y

# Test the model
model = SignalCompressor()
input_data = torch.randn(8, 20)
output = model(input_data)

print(f"Output shape: {output.shape}")
```
