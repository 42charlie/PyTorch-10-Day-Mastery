```python
import torch
import torch.nn as nn
```

# The Challenge: The Manual Gradient Tracker
*Goal:* Create a script that manually calculates a loss, triggers backpropagation.

```python
# Define model
model = nn.Linear(1, 1)

# Define inputs and targets
x = torch.tensor([[1.0]])
y = torch.tensor([[5.0]])

# Forward Pass
pred = model(x)

# Define Loss Function and Calculate Loss
criterion = nn.MSELoss()
loss = criterion(pred, y)

print(f"Current Loss: {loss.item()}")

# Gradient Check 1 (Before backward)
print(f"Gradients before backward: {model.weight.grad}")

# Backward Pass
loss.backward()

# Gradient Check 2 (After backward)
print(f"Gradients after backward: {model.weight.grad.item()}")

# The Reset
model.zero_grad()
print(f"Gradients after zero_grad(): {model.weight.grad}")
```
